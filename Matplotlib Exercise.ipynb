```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Exercise 1

*Create a figure object called fig using plt.figure()

*Use add_axes to add an axis to the figure canvas at[0,0,1,1].Call thus new axis ax.

*Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,z)
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```

# Exercise 2

Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and
[0.2,0.5,0.2,0.2]

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y)
ax2.plot(x,y,color='r')
```

# Exercise 3

Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,0.4,0.4]
Then use x,y and z arrays to recreate the plot.

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.set_title('Zoom')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('Zoom2')

ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

# Exercise 4

Use plt.subplots(nrows=1,ncols=2) to create a plot,now plot (x,y) and (x,z)on the axes. play around with line width and style.

```python
fig,axes = plt.subplots(1,2,figsize=(12,2))

axes[0].plot(x,y,color='blue') 
axes[1].plot(x,z,color='red',lw=3,ls='--')

plt.tight_layout()
```
